```python
from scripts.ambiente import Ambiente
from scripts.leitor import Leitor
from scripts.problems import Nrainhas, AlgebricFunction, FabricaDeRadios
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pickle

%load_ext autoreload
%autoreload 2
```

    NRAINHAS

```python
config = Leitor().loadConfig("data/data_fabrica_de_radios.txt")
problem = FabricaDeRadios(penality_factor=-1)
ambientes = [Ambiente(config=config,problem=problem,parallel=False,tournament_win_rate=0.9) for _ in range(9)]
for ambiente in ambientes:
    ambiente.run()
```

```python
fig, axes = plt.subplots(1,3,figsize=(12,4))
for i in range(3):
    j = 6 + i
    lucro = problem.objective_function(problem.decode(ambientes[j].elite_population[0]))
    posicao = np.array(ambientes[j].results_best).argmax()
    frase = 'Lucro de ' + str(lucro) + ',\nObtido na iteração ' + str(posicao)
    sns.lineplot(ambientes[j].results_mean,color='Blue',ax=axes[i])
    sns.lineplot(ambientes[j].results_best,color='Red',ax=axes[i]).set(title=frase)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(1,3,figsize=(12,4))
for i in range(3,6):
    j = 0
    k = i +j*3
    lucro = problem.objective_function(problem.decode(ambientes[k].elite_population[0]))
    posicao = np.array(ambientes[k].results_best).argmax()
    frase = 'Lucro de ' + str(lucro) + ',\nObtido na iteração ' + str(posicao)
    sns.lineplot(ambientes[k].results_mean,color='Blue',ax=axes[i])
    sns.lineplot(ambientes[k].results_best,color='Red',ax=axes[i]).set(title=frase)
plt.tight_layout()
```

```python
max_value = np.array([np.array(ambiente.results_best).argmax() for ambiente in ambientes])
print(max_value.mean())
print(max_value)
sns.boxplot(max_value)
```

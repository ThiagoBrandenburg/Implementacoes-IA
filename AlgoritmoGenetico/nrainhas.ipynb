```python
from scripts.ambiente import Ambiente
from scripts.leitor import Leitor
from scripts.problems import Nrainhas, AlgebricFunction, FabricaDeRadios
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pickle

%load_ext autoreload
%autoreload 2
```

```python
config = Leitor().loadConfig('data/data_nrainhas128.txt')
problem = Nrainhas(128)
```

    ESCREVER RUN

    CARREGAR RUN

```python
mem_ambiente: list[Ambiente]
with open('data/runs/nrainhas128_5000_x10.ppkl','rb') as f:
    mem_ambiente = pickle.load(f)
```

```python
fig, axes = plt.subplots(3,3,figsize=(12,12))
for i in range(3):
    for j in range(3):
        k = i +j*3
        colisoes = problem.objective_function(problem.decode(mem_ambiente[k].elite_population[0]))
        sns.lineplot(mem_ambiente[k].results_mean,color='Blue',ax=axes[i][j])
        sns.lineplot(mem_ambiente[k].results_best,color='Red',ax=axes[i][j]).set(title=str(colisoes)+' colises')
plt.tight_layout()
```

```python
for i,ambiente in enumerate(mem_ambiente):
    colisoes = problem.objective_function(problem.decode(mem_ambiente[i].elite_population[0]))
    print('Run #',i,' Best=',np.array(ambiente.results_best).argmax(), 'Colises =',colisoes)
```

```python
fig, axes = plt.subplots(3,3,figsize=(12,12))
for i in range(3):
    for j in range(3):
        k = i +j*3
        colisoes = problem.objective_function(problem.decode(ambientes[k].elite_population[0]))
        sns.lineplot(ambientes[k].results_mean,color='Blue',ax=axes[i][j])
        sns.lineplot(ambientes[k].results_best,color='Red',ax=axes[i][j]).set(title=str(colisoes)+' colises')
plt.tight_layout()
```

```python
for ambiente in ambientes:
    print('best:',ambiente.results_best)
    print('mean:',ambiente.results_mean)
    print('Solution:',ambiente.elite_population[0])
    print('Colises:',problem.objective_function(problem.decode(ambiente.elite_population[0])))
```

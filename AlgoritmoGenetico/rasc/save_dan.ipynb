    NRAINHAS

```python
config = Leitor().loadConfig('data/data_nrainhas_num8.txt')
problem = NrainhasSum()
ambientes = [
    Ambiente(
        config=config,
        problem=problem,
        parallel=False,
    ) for _ in range(4)
]
def run(ambiente:Ambiente):
    ambiente.run()
    return ambiente

ambientes = Parallel(n_jobs=-1)(
    delayed(run)(ambiente) for ambiente in ambientes
)
```

```python
sns.set_theme(style='whitegrid')
```

```python
config = Leitor().loadConfig('data/data_nrainhas_num16.txt')
problem = NrainhasSum()
ambiente = Ambiente(
        config=config,
        problem=problem,
        parallel=False)
ambiente.run()
```

```python
for i in range(2):
    for j in range(2):
        k = i + 2*j
        best_solution = problem.decode(ambiente.elite_population[0])
        fit_value = problem.fitness(best_solution)
        penality_value = problem._number_of_colisions(best_solution)
        title = '8-rainhas valorado'+'\nfuncao_fitness(Melhor) = ' + f"{fit_value:.4f}" + '\n' + 'conta_colisao(Melhor) =' + str(penality_value)
        sns.lineplot(ambiente.results_mean,linewidth=1,label='Média Fitness').set(title=title)
        sns.lineplot(ambiente.results_best,label='Melhor Fitness')
        plt.legend()
```
